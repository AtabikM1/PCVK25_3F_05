```python
# Soal no 2 Pisahkan channel
B, G, R = cv.split(img)

# Gabung channel (Red-Blue)
RB = cv.merge([R, B, np.zeros_like(R)])
cv2_imshow(RB)

# Gabung channel (Green-Blue)
GB = cv.merge([np.zeros_like(G), G, B])
cv2_imshow(GB)
```

```python
#soal no 5
canvas = np.zeros((300, 300, 3), dtype=np.uint8)

for i in range(5):
    pt1 = (np.random.randint(0, 200), np.random.randint(0, 200))
    pt2 = (pt1[0] + np.random.randint(20, 100), pt1[1] + np.random.randint(20, 100))
    color = (np.random.randint(0,256), np.random.randint(0,256), np.random.randint(0,256))
    cv.rectangle(canvas, pt1, pt2, color, -1)

cv2_imshow(canvas)
```
